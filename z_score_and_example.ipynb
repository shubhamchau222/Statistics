```python
import scipy.stats as stats
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
normal_data = np.random.normal(loc=5,
                    scale=5,  
                    size=1000)

sns.displot(normal_data, kde=True, color='green')
plt.show()

## check of data is normal distributed or not
print(stats.shapiro(normal_data))

statistic, p_value = stats.shapiro(normal_data)

# Print the result
print(f"Shapiro-Wilk statistic: {statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Check if data is normally distributed
if p_value > 0.05:
    print("The data seems to be normally distributed.")
else:
    print("The data does not seem to be normally distributed.")
```

```python
##creating log normal dist & checking if data is normaly distributed or not
normal_data = np.random.lognormal(size=1000)

sns.displot(normal_data, kde=True, color='red')
plt.show()

## check of data is normal distributed or not
print(stats.shapiro(normal_data))

statistic, p_value = stats.shapiro(normal_data)

# Print the result
print(f"Shapiro-Wilk statistic: {statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Check if data is normally distributed
if p_value > 0.05:
    print("The data seems to be normally distributed.")
else:
    print("The data does not seem to be normally distributed.")
```

# Z_score examples

## 1) **Given x_bar=70, std_dev=10, x=85 find the probability of student getting marks lessequal  than 85**

```python
mean=70
x=85
std_dev= 10

z_score = (x- mean)/std_dev
print(f"Z score is {z_score:2f}")

# find the area under curve left to 85
probability = stats.norm.cdf(z_score)

print(f"probability of getting marks <= 85 is {probability*100:4f}")
```

```python
# Plotting the normal distribution
x_vals = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
y_vals = stats.norm.pdf(x_vals, mean, std_dev)

plt.plot(x_vals, y_vals, label="Normal Distribution", color='blue')

# Highlight the area under the curve to the left of x=85
plt.fill_between(x_vals, y_vals, where=(x_vals <= x), color='red', alpha=0.5, label=f'Area <= {x}')

# Add labels and title
plt.title(f"Normal Distribution (Mean = {mean}, Std Dev = {std_dev})")
plt.xlabel('Marks')
plt.ylabel('Probability Density')

# Display the Z-score and probability
plt.text(75, 0.03, f'Z-score: {z_score:.2f}\nP(X <= {x}): {probability:.4f}', fontsize=12)
plt.text(60, 0.01, f'Area Under \nthe curve: {probability*100:.2f}\n', fontsize=10)

# Show the plot
plt.legend()
plt.grid(False)
plt.show()
```

## 2) **Given x_bar=70, std_dev=10 find the probability of student getting marks between 85 & 90**

```python
# here ub = 90 & lowerb= 85
mean=70
lb=85
ub=90
std_dev= 10

## find the z_score for ub & lb
z_score_lb = (lb-mean)/std_dev
z_score_ub = (ub-mean)/std_dev

## area left to z_score_lb
prob_lb = stats.norm.cdf(z_score_lb)
prob_ub = stats.norm.cdf(z_score_ub)

print(f" Z score of lower bound : {z_score_lb}")
print(f" Z score of upper bound : {z_score_ub}")
print(f"Area left to Lower bound : {prob_lb}")
print(f"Area left to Upper bound : {prob_ub}")

prob_lb_up = prob_ub - prob_lb
print(f"\nprobability of getting marks between 85 & 90 is {prob_lb_up*100:4f}")
```

```python
# Plotting the normal distribution
x_vals = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
y_vals = stats.norm.pdf(x_vals, mean, std_dev)
plt.plot(x_vals, y_vals, label="Normal Distribution", color='blue')
plt.grid(False)

# Highlight the area under the curve to the left of x=85
plt.fill_between(x_vals, y_vals, where=(x_vals >= lb) & (x_vals <= ub), color='red', alpha=0.5, label=f'Area <= {x}')

plt.title(f"Normal Distribution (Mean = {mean}, Std Dev = {std_dev})")
plt.xlabel('Marks')
plt.ylabel('Probability Density')

# Display the Z-score and probability
plt.text(75, 0.03, f'Z-score ub: {z_score_ub:.2f}\nP(X <= {ub}): {prob_ub:.4f}\n Z-score lb: {z_score_lb:.2f}\nP(X <= {lb}): {prob_lb:.4f} ', fontsize=12)
# plt.text(75, 0.01, f'Z-score lb: {z_score_lb:.2f}\nP(X <= {lb}): {prob_lb:.4f}', fontsize=12)

# Show the plot
plt.legend()
# plt.grid(True)
plt.show()

```
